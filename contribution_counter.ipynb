```python
import re

import pandas as pd

from issue import Issue
```

```python
TEST_FILE_PATH = "teste.csv.gzip"

sample_dataset = pd.read_csv(TEST_FILE_PATH, compression='gzip', lineterminator='\n')

issues_dict_list = sample_dataset.to_dict('records')


issues = []

for issue_dict in issues_dict_list:
	new_issue = Issue.from_dict(issue_dict)
	issues.append(new_issue)
```

```python
file = open("../log.txt")
all_text = file.read()
authors = re.findall("<author-login>(.+)</author-login>",all_text)
print(authors[:50])
```

```python
commit_no_by_author = {}

for author in authors:
	if commit_no_by_author.get(author) is None:
		commit_no_by_author[author] = 1
	else:
		commit_no_by_author[author] += 1

print(commit_no_by_author)
```
